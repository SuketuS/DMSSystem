```python
from transformers import pipeline
from sentence_transformers import SentenceTransformer
import numpy as np

# Initialize models
embedder = SentenceTransformer('all-MiniLM-L6-v2')
llm = pipeline("text-generation", model="gpt2")

# Simple document store
documents = [
    "Machine learning is a subset of artificial intelligence.",
    "Python is widely used for data science and machine learning.",
    "RAG stands for Retrieval-Augmented Generation.",
    "Transformers are powerful models for NLP tasks."
]

# Embed documents
doc_embeddings = embedder.encode(documents)

# Retrieval function
def retrieve(query, k=2):
    query_embedding = embedder.encode(query)
    scores = np.dot(doc_embeddings, query_embedding)
    top_k_idx = np.argsort(scores)[-k:][::-1]
    return [documents[i] for i in top_k_idx]

# RAG function
def rag(question):
    context = " ".join(retrieve(question))
    prompt = f"Context: {context}\n\nQuestion: {question}\nAnswer:"
    return llm(prompt, max_length=50, num_return_sequences=1)[0]['generated_text']

# Test
result = rag("What is machine learning?")
print(result)
```
